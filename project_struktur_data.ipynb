5 Nama Mahasiswa

```python
mahasiswa = ["Dode", "Manik", "Bagas", "Algian", "Dirga"]
print("List Mahasiswa: ")
print(mahasiswa)
```

Tambahan 1 Nama Baru dengan append

```python
mahasiswa.append("Kayana")
print("Setelah ditambahkan nama baru")
print(mahasiswa)
```

Hapus Satu Elemen Menggunakan Remove

```python
mahasiswa.remove:("Dirga")
print("Setelah menghapus nama Dirga")
print(mahasiswa)
```
