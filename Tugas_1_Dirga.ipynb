```python
list = []
```

Input 10 nilai


```python

for i in range(10):
  n = int(input(f"masukkan nilai mahasiswa ke-{i+1} :"))
  nilai_mahasiswa.append(n)
```

Tampilkan nilai tertinggi & terendah


```python
print("List:", nilai_mahasiswa)
print("Nilai tertinggi:", max(nilai_mahasiswa))
print("Nilai terendah:", min(nilai_mahasiswa))
```

Hitung rata-rata


```python
print("Rata-rata:", sum(nilai_mahasiswa) / len(nilai_mahasiswa))
```

Hitung jumlah mahasiswa lulus (>= 60)


```python
lulus = 0
tidak_lulus = 0

for n in nilai_mahasiswa:
    if n >= 60:
        lulus += 1
    else:
        tidak_lulus += 1

print("Jumlah lulus:", lulus)
print("Jumlah tidak lulus:", tidak_lulus)
```

Buat grafik untuk menggambarkan nilai tertinggi dan terendah


```python
import matplotlib.pyplot as plt

plt.figure()
kategori = ["Lulus", "Tidak_Lulus"]
data = [lulus, tidak_lulus]

plt.bar(kategori, data, color=["blue", "red"])
plt.title("Grafik Data Kelulusan Mahasiswa")
plt.ylabel("Jumlah Mahasiswa")
plt.ylim(0, 10)
plt.show()
```

Buat grafik untuk menggambarkan data kelulusan


```python
import matplotlib.pyplot as plt

plt.figure()
kategori = ["Tertinggi", "Terendah"]
nilai_tertinggi = max(nilai_mahasiswa)
nilai_terendah = min(nilai_mahasiswa)
data = [nilai_tertinggi, nilai_terendah]

plt.bar(kategori, data, color=["blue", "red"])
plt.title("Grafik Nilai Tertinggi dan Terendah")
plt.ylabel("Nilai")
plt.ylim(0, 100)
plt.show()
```

Analisis kompleksitas tiap operasi
